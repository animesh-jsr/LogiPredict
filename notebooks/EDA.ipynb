```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/shipments.csv')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Delivery_Delay_hr'], bins=30)

plt.title('Delivery Delay Distribution')

plt.xlabel('Delay Hours')

plt.ylabel('Frequency')

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Transport_Mode',
    y='Delivery_Delay_hr',
    data=df
)

plt.title('Delay by Transport Mode')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Weather',
    y='Delivery_Delay_hr',
    data=df
)

plt.title('Weather Impact on Delivery Delay')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Traffic_Level',
    y='Delivery_Delay_hr',
    data=df
)

plt.title('Traffic Impact on Delivery Delay')

plt.show()
```
