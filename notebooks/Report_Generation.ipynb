```python
from fpdf import FPDF

import pandas as pd
```

```python
df = pd.read_csv(
    '../reports/predicted_shipments.csv'
)

df.head()
```

```python
pdf = FPDF()

pdf.add_page()

pdf.set_font(
    "Arial",
    size=16
)
```

```python
pdf.cell(
    200,
    10,
    txt="LogiPredict Shipment Report",
    ln=True,
    align='C'
)

pdf.ln(10)
```

```python
avg_delay = round(
    df['Predicted_Delay_hr'].mean(),
    2
)

max_delay = round(
    df['Predicted_Delay_hr'].max(),
    2
)

min_delay = round(
    df['Predicted_Delay_hr'].min(),
    2
)
```

```python
pdf.set_font(
    "Arial",
    size=12
)

pdf.cell(
    200,
    10,
    txt=f"Average Predicted Delay: {avg_delay} hrs",
    ln=True
)

pdf.cell(
    200,
    10,
    txt=f"Maximum Predicted Delay: {max_delay} hrs",
    ln=True
)

pdf.cell(
    200,
    10,
    txt=f"Minimum Predicted Delay: {min_delay} hrs",
    ln=True
)
```

```python
pdf.ln(10)

pdf.set_font(
    "Arial",
    style='B',
    size=12
)

pdf.cell(
    200,
    10,
    txt="Business Recommendation:",
    ln=True
)

pdf.set_font(
    "Arial",
    size=12
)

recommendation = (
    "High-delay shipments should use "
    "alternate carriers or optimized routes."
)

pdf.multi_cell(
    0,
    10,
    recommendation
)
```

```python
pdf.output(
    '../reports/shipment_report.pdf'
)

print("PDF report generated successfully!")
```
