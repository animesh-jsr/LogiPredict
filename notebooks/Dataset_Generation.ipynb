```python
import pandas as pd
import numpy as np
```

```python
# Random seed
np.random.seed(42)

# Number of rows
n = 5000

# Categories
transport_modes = ['Truck', 'Ship', 'Air', 'Rail']
weather_conditions = ['Clear', 'Rainy', 'Storm', 'Fog']
traffic_levels = ['Low', 'Medium', 'High']

# Generate synthetic dataset
data = {
    'Distance_km': np.random.randint(50, 3000, n),

    'Transport_Mode': np.random.choice(
        transport_modes, n
    ),

    'Weather': np.random.choice(
        weather_conditions, n
    ),

    'Traffic_Level': np.random.choice(
        traffic_levels, n
    ),

    'Fuel_Price': np.random.uniform(70, 120, n),

    'Carrier_Rating': np.round(
        np.random.uniform(1, 5, n), 1
    ),

    'Warehouse_Time_hr': np.random.randint(1, 48, n)
}

# Create dataframe
df = pd.DataFrame(data)

# Generate realistic delay
delay = (
    df['Distance_km'] * 0.01
    + df['Warehouse_Time_hr'] * 0.5
    + np.where(df['Weather'] == 'Storm', 10, 0)
    + np.where(df['Traffic_Level'] == 'High', 8, 0)
    - df['Carrier_Rating'] * 2
    + np.random.normal(0, 5, n)
)

# Target variable
df['Delivery_Delay_hr'] = np.round(delay, 2)

# Show dataset
df.head()
```

```python
df.to_csv('../data/shipments.csv', index=False)

print("Dataset saved successfully!")
```
