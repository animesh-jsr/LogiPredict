```python
import pandas as pd
import numpy as np
import joblib
```

```python
model = joblib.load(
    '../models/random_forest.pkl'
)

print("Model loaded successfully!")
```

```python
sample = {
    'Distance_km': [1200],
    
    'Fuel_Price': [95],
    
    'Carrier_Rating': [4.2],
    
    'Warehouse_Time_hr': [12],

    'Transport_Mode_Rail': [0],
    'Transport_Mode_Ship': [0],
    'Transport_Mode_Truck': [1],

    'Weather_Fog': [0],
    'Weather_Rainy': [1],
    'Weather_Storm': [0],

    'Traffic_Level_Low': [0],
    'Traffic_Level_Medium': [0]
}
```

```python
sample_df = pd.DataFrame(sample)

sample_df
```

```python
prediction = model.predict(sample_df)

predicted_delay = round(
    prediction[0],
    2
)

print(
    "Predicted Delay:",
    predicted_delay,
    "hours"
)
```

```python
if predicted_delay > 20:
    
    risk = "High"
    
    recommendation = (
        "Use alternate carrier or air transport."
    )

elif predicted_delay > 10:
    
    risk = "Medium"
    
    recommendation = (
        "Monitor shipment closely."
    )

else:
    
    risk = "Low"
    
    recommendation = (
        "Shipment is on schedule."
    )

print("Risk Level:", risk)

print("Recommendation:", recommendation)
```

```python
test_shipments = pd.DataFrame({

    'Distance_km': [500, 1800, 2500],

    'Fuel_Price': [80, 100, 110],

    'Carrier_Rating': [4.5, 3.2, 2.5],

    'Warehouse_Time_hr': [5, 20, 30],

    'Transport_Mode_Rail': [0, 1, 0],

    'Transport_Mode_Ship': [0, 0, 1],

    'Transport_Mode_Truck': [1, 0, 0],

    'Weather_Fog': [0, 0, 1],

    'Weather_Rainy': [0, 1, 0],

    'Weather_Storm': [0, 0, 1],

    'Traffic_Level_Low': [1, 0, 0],

    'Traffic_Level_Medium': [0, 1, 0]
})

test_shipments
```

```python
test_shipments['Predicted_Delay_hr'] = (
    model.predict(test_shipments)
)

test_shipments
```

```python
test_shipments.to_csv(
    '../reports/predicted_shipments.csv',
    index=False
)

print("Prediction report saved!")
```
